```python
def wiki_most_watched_youtube_vids():
    #importing necessary modules
    import selenium
    from selenium import webdriver as wb
    import pandas as pd

    webD = wb.Chrome('chromedriver.exe')
    webD.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
    #initializing lists for storing data
    ranks = []
    names = []
    artists = []
    views_in_Billions = []
    upload_dates = []
    #finding all the row elements of the table
    rows = webD.find_element_by_xpath('//*[@id="mw-content-text"]/div[1]/table[1]/tbody').find_elements_by_tag_name("tr")
    # looping through all the row and extracting data
    for row in rows:
        entries = row.find_elements_by_tag_name("td")
        rank = entries[0].text
        name = entries[1].text
        artist = entries[2].text
        view = entries[3].text
        upload_date = entries[4].text
        ranks.append(rank)
        names.append(name)
        artists.append(artist)
        views_in_Billions.append(view)
        upload_dates.append(upload_date)
    webD.close()
    videos = pd.DataFrame({'Rank':ranks,'Name':names,'Artist(s)':artists,'Views (in Billions)':views_in_Billions,'Upload Date':upload_dates})
    return videos
```
