```python
def state_wise_gdp():
    #importing necessary modules
    import selenium
    from selenium import webdriver as wb
    import pandas as pd
    import time

    webD = wb.Chrome('chromedriver.exe')
    webD.get('http://statisticstimes.com/')
    economy = webD.find_elements_by_class_name('dropdown')[1].click()
    India = webD.find_elements_by_class_name('dropdown')[1].find_elements_by_tag_name('a')[1].click()
    time.sleep(5)
    webD.find_element_by_xpath('/html/body/div[2]/div[2]/div[2]/ul/li[1]/a').click()
    #initializing lists for storing data
    ranks = []
    states = []
    gdps18_19 = []
    gdps17_18 = []
    shares = []
    gdps = []
    #finding all the row elements of the table
    rows = webD.find_elements_by_tag_name('tbody')[1].find_elements_by_tag_name('tr')
    # looping through all the row and extracting data
    for row in rows:
        rank = row.find_elements_by_tag_name('td')[0].text
        state = row.find_elements_by_tag_name('td')[1].text
        gdp18_19 = row.find_elements_by_tag_name('td')[2].text
        gdp17_18 = row.find_elements_by_tag_name('td')[3].text
        share = row.find_elements_by_tag_name('td')[4].text
        gdp = row.find_elements_by_tag_name('td')[5].text
        ranks.append(rank)
        states.append(state)
        gdps18_19.append(gdp18_19)
        gdps17_18.append(gdp17_18)
        shares.append(share)
        gdps.append(gdp)
    webD.close()
    state_wise_gdp = pd.DataFrame({'Rank':ranks,'State':states,'GSDP (18-19)':gdps18_19,'GSDP (17-18)':gdps17_18,'Share':shares,'GDP - 2017 ($billion)':gdps})
    return state_wise_gdp
```
