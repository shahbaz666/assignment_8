```python
def github_trending_repos():
    #importing necessary modules
    import selenium
    from selenium import webdriver as wb
    import pandas as pd
    webD = wb.Chrome('chromedriver.exe')
    webD.get('https://github.com/')
    #click on explore followed by trending
    explore = webD.find_element_by_tag_name('nav').find_elements_by_tag_name('summary')[1].click()
    trending = webD.find_element_by_xpath('/html/body/div[1]/header/div/div[2]/nav/ul/li[4]/details/div/ul[2]/li[3]/a').click()
    titles = []
    descs = []
    languages = []
    urls = []
    #finding all the result elements on the page
    results = webD.find_elements_by_class_name('Box-row')
    # looping through all the results and extracting data
    for result in results:
        title = result.find_element_by_tag_name('h1').text
        if len(result.find_elements_by_tag_name('p'))>0:
            desc = result.find_element_by_tag_name('p').text
        else:
            desc = 'NA'
        language = result.find_elements_by_tag_name('div')[1].find_elements_by_tag_name('span')[0].text
        url = result.find_element_by_tag_name('h1').find_element_by_tag_name('a').get_property('href')
        titles.append(title)
        descs.append(desc)
        languages.append(language)
        urls.append(url)
    #from the stored URLs we will visit each page and collect count of contributors
    import numpy as np
    contr_counts = []
    for url in urls:
        webD.get(url)

        contr_objects = webD.find_elements_by_class_name('BorderGrid-row')
        contr_count = np.nan
        for obj in contr_objects:
            if obj.text.split('\n')[0].split(' ')[0]=='Contributors':#Select contributor count only if it is present in the page
                contr_count = obj.find_element_by_tag_name('h2').find_element_by_tag_name('span').text
            else:
                continue

        contr_counts.append(contr_count)
    webD.close()
    repos = pd.DataFrame({'Title':titles,'Description':descs,'Language used':languages,'Contributors count':contr_counts})
    return repos
```

```python
print(github_trending_repos())
```
